The legal states of the traffic lights can be described as $R \,?\, ¬G : G \oplus A$

<div style="display:none">
One way we could describe the next states would be the following:
$$
\begin{align}
R' &= R \oplus A \\
A' &= ¬A \\
G' &= R \land A
\end{align}
$$
</div>

It can be shown that if the traffic lights are in a legal state then $¬A \leftrightarrow G \lor (R \land ¬A)$. We can use the turnstile ($\vdash$) operator to relate the two expressions:
$$
R \,?\, ¬G : G \oplus A \vdash ¬A \leftrightarrow G \lor (R \land ¬A)
$$

> This __entailment__ is valid if any valuation that makes the left hand side true, also makes one of the conclusions on the right hand side true. In the is case the left hand side says the lights are in a legal state, and if the lights are in a legal state then the expression on the right hand side becomes true.

Assuming the following:
- If the races are fixed or the gambling houses are crooked, then the tourist trade will decline.
- If the tourist trade declines then the police force will be happy.
- The police force is never happy.

Can we conclude that the races are not fixed?

Let:
- TT: tourist trade declines
- PH: police force are happy
- RF: races are fixed
- GC: gambling houses are crooked

Thus:
- If $¬PH$ then $TT$ must be $\bot$ since $TT \rightarrow PH$.
- Since $TT$ is $\bot$ then $RF \lor GC$ must be $\bot$.
- $¬(RF \lor GC)$ must be $\top$ thus $¬RF \land ¬GC$.
- Thus we can conclude that the races are not fixed.

We can use the notation to formalize the logic into the following:
$$
RF \lor GC \rightarrow TT, \ TT \rightarrow PH,\  ¬PH \vdash ¬RF
$$

So an entailment is as follows:
$$
a_0, a_1, ..., a_n \vdash c
$$
where $a_0, a_1, ..., a_n$ are the assumptions that, if for every valuation that makes the assumptions true, makes the conclusion true.

If we have no assumptions:
$$
\vdash c
$$
Their are not assumptions, thus the left hand side is true. Thus $c$ is a totology.

If we have no conclusions:
$$
a_0, a_1, ..., a_n \vdash
$$
Then for their is not valuation that makes the righthand side true (since their is not conclusion), therfore the assumptions are inconsistent.

An empty entailment:
$$
\vdash
$$
must be invalid since the LHS is true, but their is nothing on the right therefore its not valid.

We can relate one and no conclusion entailments by observing that:
$$
a_0, a_1, ..., a_n \vdash c \quad \text{is valid iff} \quad a_0, a_1, ..., a_n ¬c \vdash
$$
if the assumptions make $c$ true, then the assumptions and $¬c$ must be inconsistent (since $a_0, ..., a_n$ make $c$ true $¬c$ will be false).

Here is another example.

Assuming:
- If I am clever then I will pass
- If I will pass then I am clever
- Either I am clever or I will pass

Conclusion:
- I am clever and I will pass

To see if this is valid lets first form an entailment
$$
C \rightarrow P, P \rightarrow C, C \lor P \vdash C \land P
$$

To prove this we need to show every valuation that makes the assumptions true make $C \land P$ true, alternativly we can say:
$$
C \rightarrow P, P \rightarrow C, C \lor P, ¬(C \land P) \vdash 
$$

If we can prove the assumptions are inconsistent then the entialment is valid. We can use resolution to do so.
