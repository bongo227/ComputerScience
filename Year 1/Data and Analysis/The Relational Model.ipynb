# The Relational Model

### Key Constraints (Constraints on Relationships)

- __Many-to-one:__ Several instances of A are related to a single instance of B
- __One-to-many:__ Each A is related to several B's
- __Many-to-many:__ Any number of A may be related to any number of B's

On an ER model, relationships that have a one to many or many to one have and arrow pointing to the "one" entity.

| Example | Type |
| :--- | :--- |
| Artist - Painting | One-to-one |
| Pens - Student | Many-to-one |
| Student - Pens | One-to-many |
| Students - Courses | Many-to-many |


### Partisipation Constraints
- __Total Partisipation:__ Every instance of an entity E has atleast one relationship instance R.
- __Partial Partisipation:__ Some instances of an entity E may have no relationship instance R.

On an ER model, total partisipation is represented by a double line.

| Example | Type |
| :--- | :--- |
| Student - Language | Total partisipation |
| Course - Exam | Partial partisipation |

### The Relation Model
- __Relations:__ The compination of a schema and an instance.
- __Schema:__ List of fields (attributes).
- __Instace:__ A record/row/tuple with all the data.
- __Arity:__ Number of fields.
- __Cardinality:__ Number of rows.
- __Foreign key:__ Used to link tables in a relational database by storing the primary key of another instance.

The relational model is much closer to the actual implentation of the database. It uses tables with foreign keys to link instances. SQL is a standard language for interacting with a relational database. Data definition language (DDL) is the portion of SQL used to declare the schema.

#### Types
- `INTEGER` for integer values.
- `FLOAT` for floatinig point values.
- `VARCHAR(n)` strings of length `n`.

#### Creating a table
```
CREATE TABLE Takes (
    uun VARCHAR(8),
    code VARCHAR(20),
    mark INTEGER,
    PRIMARY KEY (uun, code),
    FOREIGN KEY (uun) REFERENCES Student,
    FOREIGN KEY (code) REFERENCES Course,
```
Here we created a join table that links a student by their `uun` with a course by its `code`.
