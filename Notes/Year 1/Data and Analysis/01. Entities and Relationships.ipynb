# Entities and Realationships

### Database design
1. __Requirement anaylis:__
    Understand what data is to be stored and the operations required.
2. __Conceptual design:__
    Developing a high level description (use of an ER data model).
3. __Logical design:__
    Implementing the conceptual model to produce a logical schema.

### ER Data Model

__ER Data Model:__ A graphical notation to organize entities and relationships that is easy to map to logical data models.

__Entity:__ A king of thing that me model in a database.

__Entity instance:__ One specific example of an entity.

__Entity set:__ A set of entitys.

__Attrubibutes:__ list the infomation we recored about an entity, they have domains such as strings up to n characters etc.

__Superkey:__ A set of attributes that uniquly identity an entitiy instance.

__Key:__ The smallest set of attributes that uniquly identity an entitiy instance.

__Relationship:__ An assosiation between entitiys

__Relationship Instance:__ A pair or tuple of primary keys

__Relationship Sets:__ A set of relationship instances

### Drawing an ER diagram
- Boxes represent entitys
- Ovals list attributes 
    - Underlined attributes are part of a primary key
- Diamond represents a relationship
    - Relationships may have attributes
