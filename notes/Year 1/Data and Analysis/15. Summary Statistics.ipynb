### Why statistics?

Their are many reasons to _analyse_ data, for example:

- To discover inplicit structure / patterns in experimental data which might suggest new models or experiments.
- To confirm or refute a hypthesis about the data.

Mathmatical statistics allow us to:

- Sensitively detect information not apparent withing the data.
- Determine wether an apparent feature actually exsists.

__Qualitative:__ yes/no; true/false; town of residence; course grade

__Quantative:__ date; population; length; weight

### Normal distribution

Many quantitve data sets form a normal distribution, which is a symetrical bell-shapped frequency curve. It has two parameters, $\mu$ (mean) which is the center of the curve and $\sigma$ (standard deviation) the spread of the curve, also the length between mu and the inflection point.



### Statistics

A _statistic_ is a single value computed from the data, which captures some property of the data.

- __Mode:__ The most common value. Data may be bi-modal (two modes) or multimodal (more than two modes). Not usefull with quantitive data sets since values are unlikey to appear twice.
- __Median:__ The middle (or average of the middle most) value in a sorted set of data. The median is usefull when their are extream outliers at either end, or the data is cut short.
- __Mean:__ The total of a data set divided by the number of elements. Uses all the data, so its a genuine summary of the data but may not be the correct choice since it can be distored by exteam outliers.
- __Standard Deviation:__ $\sigma = \sqrt{\dfrac{1}{N}\displaystyle\sum^N_{i=1}{(x_i - \mu)^2}} = \sqrt{\left(\dfrac{1}{N}\displaystyle\sum^N_{i=1}{x_i^2}\right) - \mu^2}$. The measure of spread within a dataset.



### Sampling

Most of the time it impracticle to measure the entire population, thus a sample of the population is gathered which 
is used to estimate the statitics of the population.

- Samples must be chosen _at random_.
- Samples must be _as large as practally possible_ given the con straints of gathering, storing and calculating the data.

These make the sample more _representative_ of the population they were sampled from, reducing bias.

To estimate the population parameters, we use _unbiased estimators_. An estimator is unbiased if the expectation of the estimator is the populator parameter, the values of an estimator are evenly distributed around the population parameter.

The sample mean is the best unbaised estimator for the population mean, however the sample standard deviation is not. The standard deviation requires the _bassel correction_: $s = \sqrt{\dfrac{1}{n - 1}\displaystyle\sum^n_{i=1}{(x_i - \mu)^2}} = \sqrt{\left(\dfrac{1}{n - 1}\displaystyle\sum^n_{i=1}{x_i^2}\right) - \mu^2}$


### Correlation

If we have data in two dimensions, we can find relationships between data sets. If their is no relationship, the variables are said to be _independent_, if there is a relationship the variables are _correlated_. Two variables are _causally_ connected if variation in the first causes variation in the second. 

The _correlation coefficent_ is a measure of how closelly one set of data values are correlated with another.
$$
\rho_{x, y} = \dfrac{\sum^N_{i=1}{(x_i - \mu_x)(y_i - \mu_y)}}{N \sigma_x \sigma_y}
$$

- If $\rho_{x, y}$ is close to 0, then their is no correlation.
- If $\rho_{x, y}$ is close to -1, then their is negative correlation.
- If $\rho_{x, y}$ is close to +1, then their is possitive correlation.

The estimate $r_{x, y}$ of the correlation coefficent is:
$$
\rho_{x, y} = \dfrac{\sum^n_{i=1}{(x_i - \bar{x})(y_i - \bar{y})}}{(n-1) \sigma_x \sigma_y}
$$
