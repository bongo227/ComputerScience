## Cut rule
The cut rule allows us to combine statements if an expression appears on the hyptothesis of one statement, and the conclusion of another.

Here is an example:
$$
\frac
{F \rightarrow A, B \quad F', A \rightarrow B'}
{F, F' \rightarrow B, B'}
$$

## More rules

Here are some more logic rules

$$
\frac{A \vdash X \quad A \vdash Y}{A \vdash X \land Y} \quad (\land)
$$

$$
\frac{A, X \vdash Z \quad A, Y \vdash Z}{A, X \lor Y \vdash Z} \quad (\lor)
$$

$$
\frac{A \vdash X \quad A \vdash Y}{A \vdash X \rightarrow Y} \quad (\rightarrow)
$$

These rules should have a double line since they work in both ways, going down __introduces the connective__ and going up __eliminates the connective__.

For example if from A we can derive X and from A we can derive Y (as in the first rule), then it follows that we derive X and Y from A. Likewise if we can derived X and Y from A, then we can derive X from A and Y from A. 

## Example proof

$$
\overline{A \rightarrow (B \rightarrow C) \vdash A \rightarrow (B \rightarrow C)}^{\quad\displaystyle(I)}
\\
\overline{A \rightarrow (B \rightarrow C), A \vdash B \rightarrow C}^{\quad\displaystyle(\rightarrow^-)}
\\
\overline{A \rightarrow (B \rightarrow C), B, A \vdash C}^{\quad\displaystyle(\rightarrow^-)}
\\
\overline{A \rightarrow (B \rightarrow C), B \vdash (A \rightarrow C)}^{\quad\displaystyle(\rightarrow^+)}
\\
\overline{A \rightarrow (B \rightarrow C) \vdash B \rightarrow (A \rightarrow C)}^{\quad\displaystyle(\rightarrow^+)}
$$

## Gentzen's Rules

A sequent, $A \vdash B$ is valid iff
- every expression in $A$ is true
- some expression in $B$ is true

$$
\overline{\Gamma, A \vdash \Delta, A}^{\quad\displaystyle(I)}
$$

$$
\frac{\Gamma, A, B \vdash \Delta}{\Gamma, A \land B \vdash \Delta}
\quad (\land L)
$$

$$
\frac{\Gamma, A \vdash \Delta \quad \Gamma, B \vdash \Delta}{\Gamma, A \lor B \vdash \Delta}
\quad (\lor L)
$$

$$
\frac{\Gamma \vdash A, B, \Delta}{\Gamma \vdash A \land B, \Delta}
\quad (\lor R)
$$

$$
\frac{\Gamma \vdash A, \Delta \quad \Gamma \vdash B, \Delta}{\Gamma \vdash A \land B, \Delta}
\quad (\land R)
$$

$$
\frac{\Gamma \vdash A, \Delta \quad \Gamma \vdash B, \Delta}{\Gamma, A \rightarrow B \vdash \Delta}
\quad (\rightarrow L)
$$

$$
\frac{\Gamma, A \vdash B, \Delta}{\Gamma \vdash A \rightarrow B, \Delta}
\quad (\rightarrow R)
$$

$$
\frac{\Gamma \vdash A,\Delta}{\Gamma, ¬A \vdash \Delta}
\quad (¬L)
$$

$$
\frac{\Gamma, A \vdash \Delta}{\Gamma \vdash ¬A, \Delta}
\quad (¬R)
$$

Going back to the previous proof, we can now remove the need for a cut by using the rules above

$$
\overline{A, B \vdash A, B}^{\quad\displaystyle(I)} \\
\overline{A \land B \vdash A , B}^{\quad\displaystyle(\land L)} \\
\overline{A \land B \vdash A \lor B}^{\quad\displaystyle(\lor R)}\\
$$

Lets look at a more complex proof, say we want to prove:
$$
A \rightarrow (B \rightarrow C) \vdash B \rightarrow (A \rightarrow C)
$$

First we find the a conclusion of one of the rules that matches the pattern of the expression which is in this case is $(\rightarrow R)$

- $\Gamma$ is $A \rightarrow (B \rightarrow C)$
- $\Delta$ is empty
- $A$ is $B$
- $B$ is $A\rightarrow C$

Then we can use the top half of the expression to simplify it:
$\Gamma, A \vdash B, \Gamma$ is the top half of $(\rightarrow R)$ so the top half becomes:
$$
A \rightarrow (B \rightarrow C) \vdash A, B \rightarrow C
$$

Applying the same rule again gives us:
$$
A \rightarrow (B \rightarrow C), B, A \vdash C
$$

We have an implication on the left so we want to apply the $(\rightarrow L)$ rule.
$$
B, A \vdash A, C \quad B \rightarrow C, B, A \vdash C
$$

The expression on the left is the imediant rule, so it is true. The expression on the right needs one final application of the $(\rightarrow L)$ rule which produces:
$$
B, A \vdash B, C \quad C, B, A \vdash C
$$

which are both match imediant rule so we have proved the original expression.
